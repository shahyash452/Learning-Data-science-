```python
import pandas as pd
```

```python
df=pd.read_csv('Titanic.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
new_df = df
new_df
```

```python
new_df.drop(columns = ['Ticket', 'Cabin', 'Name'], axis = 1, inplace = True)
```

```python
new_df.head()
```

```python
new_df.isnull().sum()
```

```python
new_df['Age'] = df.Age.fillna(df.Age.mean())
new_df['Fare'] = df.Fare.fillna(df.Fare.mean())
```

```python
new_df.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder 
ohe = OneHotEncoder()
```

```python
sex_encoded = ohe.fit_transform(new_df[['Sex']]).toarray()
new_df[['female','male']] = sex_encoded
```

```python
embarked_encoded = ohe.fit_transform(new_df[['Embarked']]).toarray()
new_df[['C','Q','S']] = embarked_encoded
```

```python
new_df.head()
```

```python
new_df.drop(columns = ['Sex', 'Embarked'], axis = 1, inplace = True)
```

```python
new_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
t_df = new_df[['Pclass','Age','SibSp','Parch','Fare','female','male','C','Q','S']]
```

```python
t_df.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(t_df, new_df.Survived, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```
