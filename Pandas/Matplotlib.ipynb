```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
days = [1,2,3,4,5,6,7]
min_temp = [12,13,14,15,16,17,18]
max_temp = [24,25,26,27,28,29,30]
avg_temp = [16,17,18,19,20,21,22]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days,min_temp, label='Minimum')
plt.plot(days,max_temp, label='Maximum')
plt.plot(days,avg_temp, label='Average')

plt.legend(loc='lower right')
plt.grid()
```

```python
company = ['HUL', 'Reliance' , 'ITC']
revenue = [1000,2000,1500]
profit =  [100,200,150]
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.xticks(ypos,company)
plt.xlabel('Company')
plt.ylabel('Revenue (in Cr)')
plt.title('Finance Chart')

plt.bar(ypos-0.2,revenue,width=0.4, label='Revenue')
plt.bar(ypos+0.2,profit,width=0.4, label='Profit')
plt.legend()
```

```python
exp_val = [1500,2500,3500,3200]
exp_labels = ['House Rent', 'Phone', 'Other', 'Electronics']
```

```python
plt.pie(exp_val, labels= exp_labels, autopct='%0.1f%%', shadow=True, explode=[0,0.1,0,0])
plt.legend()
plt.savefig('pie_chart.pdf', bbox_inches='tight')
```
