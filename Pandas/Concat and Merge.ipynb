```python
import pandas as pd
india_weather = pd.DataFrame({
    'city' : ['Mumbai', 'Ahmedabad', 'Nadiad'],
    'temperature' : [32,35,28],
    'humidity' : [80,60,70]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    'city' : ['New York', 'California', 'Texas'],
    'temperature' : [22,15,28],
    'humidity' : [70,50,90]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather])
df
```

```python
df = pd.concat([india_weather, us_weather],ignore_index=True)
df
```

```python
df = pd.concat([india_weather, us_weather], keys=['India','USA'])
df
```

```python
df1 = pd.DataFrame({
    'city' : ['Nadiad','Anand','Ahmedabad'],
    'temperature' : [18,19,20],
    'windspeed': [20,21,22]
})
df1
```

```python
df2 = pd.DataFrame({
    'city' : ['Anand','Nadiad','Ahmedabad','Vadodara'],
    'windspeed' : [28,29,10,11],
    'temperature' : [38,39,40,41]
})
df2
```

```python
# df3 = pd.merge(df1,df2,on='city',how='inner') #By default
df3 = pd.merge(df1,df2,on='city',how='outer')
df3
# df3 = pd.merge(df1,df2,on='city',how='right')
# df3 = pd.merge(df1,df2,on='city',how='left')
```

```python
df4 = pd.merge(df1,df2,on='city',how='outer',suffixes=('_left','_right'))
df4
```
